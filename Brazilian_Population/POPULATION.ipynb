       
## Estimativas da população residente nos municípios brasileiros com data de referência em  1º de Julho do ano selecionado      

#### Script to join populational data of Brazil. The dataset is from Fonte: IBGE. Diretoria de Pesquisas - DPE - Coordenação de População e Indicadores Sociais - COPIS.		


Links: 1. https://downloads.ibge.gov.br/downloads_estatisticas.htm

```python
import pandas as pd
from pandas import DataFrame
import numpy as np
import csv
import sys
import glob
```

```python
# get data file names
path =r'C:\Users\juliane.oliveira\Dropbox\CIDACS\População\POP_python\files'
filenames = glob.glob(path + "/*.xls")
```

```python
lst_df = []
for f in filenames:
    df = pd.read_excel(f)
    lst_df.append(df)
```

```python
df.head()
```

```python
df = pd.read_excel(filenames[8])
POP = df
for i in range(8):
    POP=POP.merge(lst_df[i], on='cod_full', how='left') 
```

```python
POP.head()
```

```python
POP.to_excel('Pop_data.xls')
```

```python
Pop_data = pd.read_excel('Pop_data.xls')
Pop_data.info()
```

```python
pop = Pop_data.filter(['UF_x','COD_UF_x', 'NOME_MUNI_x',
                       'cod_full','POP2018','POP2017',
                      'POP2016','POP2015','POP2014',
                      'POP2013','POP2012','POP2011',
                      'POP2010'], axis=1)
```

```python
null_columns= pop.columns[pop.isnull().any()]
s= pop[null_columns].isnull().sum()
df_null = s.reset_index()
df_null.columns = ['Variable','Null_Sum']

df_null
```

```python
pop['codmunres'] = pop['cod_full'].astype(str).str[:-1].astype(np.int64)
```

```python
pop.info()
```

```python
pop = pop.rename(index=str, columns={"UF_x": "UF", "COD_UF_x": "COD_UF", 
                                   "NOME_MUNI_x": "NOME_MUNI", 
                                    "cod_full": "cod_full_muni"})
```

```python
pop.to_csv('2019_01_31_Populacional_eco.csv')
```

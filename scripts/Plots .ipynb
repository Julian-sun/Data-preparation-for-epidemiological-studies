```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import *
import os
```

```python
series = pd.read_csv("/Users/julianeoliveira/Desktop/Modelling TaskForce Arboviruses/Data/data.csv")
series.head()
```

```python
dt = pd.date_range(start='2014-12-29', periods = 156, freq = 'W-MON')
```

```python
series['date'] = dt
```

```python
fig, [ax1,ax2, ax3] = plt.subplots(nrows=3, figsize=(20,10), sharex=True, sharey=False)

plt.subplots_adjust(wspace=1.5)

ax1.plot(series.date,series['cases_dengue'], label='Dengue',color='green')
ax1.set_ylabel('Nº casos',fontsize=18)
ax1.legend(loc="upper left")

ax2.plot(series.date,series['cases_zika'], label='Zika',color='orange')
ax2.set_ylabel('Nº casos',fontsize=18)
ax2.legend(loc="upper left")

ax3.plot(series.date,series['cases_chik'], label='Chikungunya')
ax3.set_ylabel('Nº casos',fontsize=18)
ax3.legend(loc="upper left")
    
         
plt.savefig('arbo_series.png')
    
   
```
